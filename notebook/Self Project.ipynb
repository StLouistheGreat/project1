We could expect that the mean number of crashes over the weekends is bigger than during labour days. 
Make a two sample t-test to evaluate this hypothesis following the steps

```python
from scipy import stats

# Categorize crashes into weekends and weekdays
cleaned_data['Is Weekend'] = cleaned_data['Day of Week'].isin(['Saturday', 'Sunday'])

# Group data by date and weekend status to get daily crash counts
daily_crashes = cleaned_data.groupby([cleaned_data['Crash Date/Time'].dt.date, 'Is Weekend']).size().reset_index(name='Counts')

# Separate the weekend and weekday data
weekend_data = daily_crashes[daily_crashes['Is Weekend'] == True]['Counts']
weekday_data = daily_crashes[daily_crashes['Is Weekend'] == False]['Counts']

# Perform the two-sample t-test
t_stat, p_value = stats.ttest_ind(weekend_data, weekday_data)

t_stat, p_value
```

p-value < 0.05: The result is statistically significant, and we reject the null hypothesis.
There is a statistically significant difference in the mean number of crashes between weekends and weekdays.
Since the t-statistic is negative, it suggests that the mean number of crashes on weekends is lower than on weekdays.
These results indicate that the mean number of crashes is significantly different between weekends and weekdays, with weekdays having a higher mean number of crashes

Let us try and visualise this

```python


# Categorize crashes into weekends and weekdays
cleaned_data['Is Weekend'] = cleaned_data['Day of Week'].isin(['Saturday', 'Sunday'])

# Group data by date and weekend status to get daily crash counts
daily_crashes = cleaned_data.groupby([cleaned_data['Crash Date/Time'].dt.date, 'Is Weekend']).size().reset_index(name='Counts')

# Plotting box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Is Weekend', y='Counts', data=daily_crashes, palette='viridis')
plt.xlabel('Is Weekend')
plt.ylabel('Number of Crashes')
plt.title('Comparison of Daily Crash Counts: Weekdays vs. Weekends')

# Calculate and plot means
mean_weekend = daily_crashes[daily_crashes['Is Weekend'] == True]['Counts'].mean()
mean_weekday = daily_crashes[daily_crashes['Is Weekend'] == False]['Counts'].mean()
plt.axhline(mean_weekend, color='blue', linestyle='--', label=f'Mean Weekend: {mean_weekend:.2f}')
plt.axhline(mean_weekday, color='orange', linestyle='--', label=f'Mean Weekday: {mean_weekday:.2f}')

plt.legend()
plt.show()
```

Box Plot: The box plot shows the distribution of daily crash counts for weekends and weekdays.

Box: Represents the interquartile range (IQR), containing the middle 50% of the data.
Whiskers: Extend to the minimum and maximum values within 1.5 times the IQR.
Dots: Represent outliers beyond the whiskers.
Mean Overlay: The dashed lines show the mean number of crashes for weekends and weekdays.

Blue Line: Mean number of crashes on weekends.
Orange Line: Mean number of crashes on weekdays.
This visualization helps to see not only the difference in the means but also the spread and variability of the daily crash counts for both groups.
